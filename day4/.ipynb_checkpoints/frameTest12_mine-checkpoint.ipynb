```python
df = pd.read_excel(r'hotel.xlsx', index_col=0)
df
```

```python
# 호텔 등급 별 가격의 합? 의미는 없음
g = df.groupby('grade')
type(g)
```

```python
# group(집계) 관련멤버함수(sum, count, max, min, mean, agg)
g.sum()['price'] # 집계 가능한 모든 column (e.g., int, float)
g.sum()[['price']]  # => data frame 뽑아냄
```

```python
g.max()
```

```python
g.count()
```

```python
grp = g.agg(['sum', 'count', 'mean'])['price']
```

```python
grp.columns = ['가격합계', '갯수', '가격평균']
grp.index.name = '호텔등급'
grp
```

```python
grp.rename(columns={'갯수':'등급갯수'}, inplace=True)          # column 명 또는 index 명 수정시 사용
grp
```

```python
grp.rename(index={5:'5등급'}, inplace=True)
grp
```

```python
grp.rename(index=(lambda v: str(v)+'등급'))
```

```python
grp.reset_index(inplace=True)
grp
```

```python
grp.set_index('호텔등급', inplace=True)
grp
```

```python
d2014 = pd.read_csv(r'2014년 졸음운전 교통사고.csv', engine='python')
d2015 = pd.read_csv(r'2015년 졸음운전 교통사고.csv', engine='python')
d2016 = pd.read_csv(r'2016년 졸음운전 교통사고.csv', engine='python')
```

```python
whole_data = pd.concat([d2014, d2015, d2016])
```

```python
# df = pd.concat( [df14, df15, df16], axis=0 )
# # df.set_index('구분')

# df['년도'] = df['구분'].str.slice(0,5)
# df['월'] = df['구분'].str.slice(5,8)
# df.drop(columns=['구분'], inplace=True)
# df.set_index(['년도','월'],inplace=True)
# df

whole_data
pivot_data = whole_data
pivot_data['년'] = pivot_data['구분'].apply(lambda v: v.split('년')[0])
pivot_data['월'] = pivot_data['구분'].apply(lambda v: int(v.split('년')[1].split('월')[0]))
```

```python
pivot_data
```

```python
# 1-1
pivot_data2 = pivot_data.pivot_table(index=['년', '월'])
pivot_data2
```

```python
# 1-2
pivot_data3 = pivot_data.pivot_table(index=['월', '년'])
pivot_data3
```

```python
# 2
rate = pivot_data2.loc['2016'].sum()['사망(명)'] / pivot_data2.loc['2016'].sum()['사고(건)'] * 100
print("사고대비사망률: %f" % rate)
```

```python
# 3
pivot_data2.loc['2014'][['사망(명)', '부상(명)']].plot(kind='bar')
plt.show()
```

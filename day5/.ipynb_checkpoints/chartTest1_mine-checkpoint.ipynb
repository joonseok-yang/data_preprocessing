```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.plot(x,y1)
plt.figure(figsize=[3,3])        # chart를 그리기 위한 공간을 만든다! 자동으로 추가되어 있다고 보면 된다.
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.subplot(2,2,1)         # plot창을 2 x 2로 분할하라
plt.plot(x,y1)             # [[1,2][3,4]] index 할당 됨

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.subplot(1,3,1)
plt.plot(x,y1)

plt.subplot(1,3,2)
plt.plot(x,y2)

plt.subplot(1,3,3)
plt.plot(x,y3)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1)
plt.show()
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html   => document
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1, 'ro-', x, y2, 'b>--', x, y3, 'gs-.')           # red, circle marker, dash-dot line style
plt.legend(['국어', '영어', '수학'])
plt.show()
```

```python
x = ['홍길동', '이순신', '임꺽정', '김철수', '이황']
y1 = np.random.randint(30, 100, size=5)

plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')
plt.grid(True)
plt.plot(x, y1, 'ro--')
plt.legend(['국어'])
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.plot(x, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10점')
plt.text(2, 14, '14점')
plt.show()
```

```python
xname = ['홍길동', '이순신', '임꺽정', '김철수', '이황']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.plot(xname, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10점')
plt.text(2, 14, '14점')
plt.show()
```

```python
xname = ['홍길동', '이순신', '임꺽정', '김철수', '이황']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.xticks(x, xname, rotation='90')
plt.plot(x, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10점')
plt.text(2, 14, '14점')
plt.show()
```

```python
xname = ['홍길동', '이순신', '임꺽정', '김철수', '이황']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.xticks(x, xname, rotation='45')
plt.plot(x, y1, 'go-', alpha=0.2)
for x, y in zip(x, y1):
    plt.text(x - 0.1, y - 0.5, '%d점' % y)
plt.show()
```

```python
x = ['홍길동','이순신','임꺽정','김철수','이황']
y1 = np.random.randint(30,100,size=5)
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')
plt.grid(True)
plt.plot(x,y1,'ro--')
plt.legend(['국어'])
plt.show()
```

```python
xname = ['홍길동','이순신','임꺽정','김철수','이황']
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
plt.xticks( x, xname,rotation='45' )
plt.plot(x,y1,'ro-',alpha=0.5)
for n1,n2 in zip(x,y1):
    plt.text(n1+0.1,n2,'%d점'%n2 )    
# plt.text(1,10,'10점')
# plt.text(2,14,'14점')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
warnings.simplefilter('ignore')
```

```python
kor=[50, 20, 80, 90, 50, 30, 70, 80, 90, 50]
eng=[20, 90, 30, 10, 70, 60, 20, 50, 40, 50]
math=[20, 30, 50, 20, 25, 65, 100, 23, 40, 47]
name=['홍길동', '이순신', '임꺽정', '이이', '이황', '김철수1', '김철수2', '김철수3', '김철수4', '김철수5']


df = pd.DataFrame({'국어': kor, '영어': eng, '수학': math})
df.index = name
df
```

```python
df_kor = df['국어']
df_kor.plot(kind='bar')
# plt.bar(df_kor.index, df_kor[])
# plt.show()
```

```python
df_eng = df['영어']
df_eng.plot(kind='bar')
```

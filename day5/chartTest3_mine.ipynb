```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
warnings.simplefilter('ignore')
```

```python
x = ['홍길동', '이순신', '임꺽정', '이이']
y = [90, 77, 50, 80]
plt.bar(x, y)
plt.show()
```

```python
x = ['홍길동', '이순신', '임꺽정', '이이']
y = [90, 77, 50, 80]
plt.barh(x, y)
plt.show()
```

```python
x = ['홍길동', '이순신', '임꺽정', '이이']
kor = [90, 77, 50, 80]
eng = [60, 20, 50, 20]
plt.bar(x, kor)
plt.bar(x, eng)
plt.legend(['국어', '영어'])
plt.show()
```

```python
x = ['홍길동', '이순신', '임꺽정', '이이']
kor = [90, 77, 50, 80]
eng = [60, 20, 50, 20]
plt.bar(x, kor, width=0.4)
plt.bar(x, eng, width=0.4, bottom=kor)       # stack bar...
plt.legend(['국어', '영어'])
plt.show()
```

```python
xname = ['홍길동', '이순신', '임꺽정', '이이']
x = [1, 2, 3 ,4]
kor = [90, 77, 50, 80]
eng = [60, 20, 50, 20]
plt.xticks(x, xname)
plt.bar(x, kor, width=0.4, alpha=0.5)
plt.bar(x, eng, width=0.4, alpha=0.5, bottom=kor)       # stack bar...
plt.legend(['국어', '영어'])
for x, k, e in zip(x, kor, eng):
    plt.text(x, k/2, '%d점' % k, va='center', ha='center')
    plt.text(x, k+(e/2), '%d점' % e, va='center', ha='center')
plt.show()
```

```python
xname = ['홍길동', '이순신', '임꺽정', '이이']
x = pd.Series([1, 2, 3 ,4])     
kor = [90, 77, 50, 80]
eng = [60, 20, 50, 20]

plt.xticks(x+0.2, xname)
plt.bar(x, kor, width=0.4, alpha=0.5)
plt.bar(x + 0.4, eng, width=0.4, alpha=0.5) # x를 series 로 만들어야 연산 가능
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.subplot(2,2,1)         # plot 2 x 2 
plt.plot(x,y1)             # [[1,2][3,4]] index  

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.xlabel('x')
plt.ylabel('y')
plt.title('')
plt.grid(True)
plt.plot(x,y1)
plt.show()
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html   => document
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]
y2 = [5, 8, 10, 25, 20]
y3 = [30, 5, 20, 11, 15]

plt.xlabel('x')
plt.ylabel('y')
plt.title('')
plt.grid(True)
plt.plot(x,y1, 'ro-', x, y2, 'b>--', x, y3, 'gs-.')           # red, circle marker, dash-dot line style
plt.legend(['', '', ''])
plt.show()
```

```python
x = ['', '', '', '', '']
y1 = np.random.randint(30, 100, size=5)

plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.grid(True)
plt.plot(x, y1, 'ro--')
plt.legend([''])
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.plot(x, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10')
plt.text(2, 14, '14')
plt.show()
```

```python
xname = ['', '', '', '', '']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.plot(xname, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10')
plt.text(2, 14, '14')
plt.show()
```

```python
xname = ['', '', '', '', '']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.xticks(x, xname, rotation='90')
plt.plot(x, y1, 'go-', alpha=0.2)
plt.text(1, 10, '10')
plt.text(2, 14, '14')
plt.show()
```

```python
xname = ['', '', '', '', '']
x = [1, 2, 3, 4, 5]
y1 = [10, 14, 19, 20, 25]

plt.xticks(x, xname, rotation='45')
plt.plot(x, y1, 'go-', alpha=0.2)
for x, y in zip(x, y1):
    plt.text(x - 0.1, y - 0.5, '%d' % y)
plt.show()
```

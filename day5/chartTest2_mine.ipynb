```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
warnings.simplefilter('ignore')
```

```python
height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter( height, weight )
plt.show()
```

```python
height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter([str(n) for n in height], weight, c=['r', 'g', 'b', 'k', 'm', 'y', 'c']
           , s=[200, 150, 10, 20, 400, 44, 230], alpha=0.5)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.grid(True)
plt.show()
```

```python
bsize=[]
bcolor=[]
btext=[]
def obesity( h, w):
    stdW = (h-100)*0.85
    biman = w/stdW*100
    print(biman)
    if biman<=90:
        bsize.append(100)
        bcolor.append('y')
        btext.append('저체중')
    elif  90 < biman<=110:
        bsize.append(200)
        bcolor.append('b')
        btext.append('정상')
    elif  110 < biman<=120:
        bsize.append(300)
        bcolor.append('c')
        btext.append('과체중')
    else:
        bsize.append(400)
        bcolor.append('r')
        btext.append('비만')        
```

```python
for h,w in zip(height,weight):
    obesity(h,w)
```

```python
print( bsize )
print( bcolor )
print( btext)
```

```python
def fn(v):
    if v <= 90:
        return "저체중"
    elif 90 < v <= 110:
        return "정상"
    elif 110 < v <= 120:
        return "과체중"
    else:
        return "비만"
obesity_degree = [ (w / ((h - 100) * 0.85)) * 100 for h, w in zip(height, weight)]
obesity_label = [fn(x) for x in obesity_degree]
print(obesity_degree)
print(obesity_label)
```

```python
print([(x//10) for x in obesity_degree])
plt.figure(figsize=[10, 10])
plt.scatter([str(n) for n in height], weight, s=[(x//10)**3.3 for x in obesity_degree])
# plt.text(str(height[0]), weight[0], 'hello')
for x, y, z in zip(height, weight, obesity_label):
    plt.text(str(x), y, z)
    
plt.show()
```

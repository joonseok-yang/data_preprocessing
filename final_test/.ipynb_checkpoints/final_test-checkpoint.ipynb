```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings
import seaborn as sns

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
warnings.simplefilter('ignore')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# 1
deck_remove = titanic.drop('deck', axis=1)
deck_remove
nan_remove = deck_remove.dropna(subset=['age'])
titanic = nan_remove
titanic
```

```python
# 2
titanic.groupby('survived')[['survived']].count()
```

```python
# 3

# ìƒì¡´ìž ìˆ˜
surv = titanic[titanic['survived'] == 1].groupby('pclass')[['survived']].count()
# íƒ‘ìŠ¹ìž ìˆ˜
tot = titanic.groupby('pclass')[['survived']].count()
survived_rate = surv / tot * 100
# ìƒì¡´ë¥ 
survived_rate
```

```python
# 4 ê°€ì¡± ìˆ˜ ë³„ ìƒì¡´ë¥ 

# ìƒì¡´ìž ìˆ˜
surv = titanic[titanic['survived'] == 1].groupby('sibsp')[['survived']].count()
# íƒ‘ìŠ¹ìž ìˆ˜
tot = titanic.groupby('sibsp')[['survived']].count()
survived_rate = surv / tot * 100
# ìƒì¡´ë¥ 
survived_rate[1:].fillna(0)
```

```python
# 4 ê°€ì¡±í•¨ê»˜ ìƒì¡´ë¥ 
surv = titanic[(titanic['survived'] == 1) & (titanic['sibsp'] > 0)].groupby('sibsp')[['survived']].count()

tot = titanic[(titanic['sibsp'] > 0)].groupby('sibsp')[['survived']].count()
with_family_survival_rate = surv.sum() / tot.sum() *100
print("ê°€ì¡±ê³¼ í•¨ê»˜ íƒ‘ìŠ¹í•œ ê²½ìš° ìƒì¡´ë¥ : %f %%" % with_family_survival_rate)
```

```python
# 5
# ìƒì¡´ìž ìˆ˜
surv = titanic[titanic['survived'] == 1].groupby('alone')['survived'].count()
# íƒ‘ìŠ¹ìž ìˆ˜
tot = titanic.groupby('alone')['survived'].count()
survived_rate = surv / tot * 100
# ìƒì¡´ë¥ 
survived_rate[True]
```

```python
# 6
# ìƒì¡´ìž ìˆ˜
surv = titanic[titanic['survived'] == 1].groupby('sex')['survived'].count()
# íƒ‘ìŠ¹ìž ìˆ˜
tot = titanic.groupby('sex')['survived'].count()
survived_rate = surv / tot * 100
# ìƒì¡´ë¥ 
survived_rate
```

```python
# 7
titanic.groupby('pclass')['fare'].mean().plot(kind='bar')
plt.show()
```

```python
# 8
def age_category(age):
    if 1 <= age < 15:
        return "ë¯¸ì„±ë…„ìž"
    elif 15 <= age < 25:
        return 'ì²­ë…„'
    elif 25 <= age < 35:
        return 'ì¤‘ë…„'
    elif 35 <= age < 60:
        return 'ìž¥ë…„'
    else:
        return 'ë…¸ë…„'

titanic['age_category'] = titanic['age'].apply(age_category)
titanic
```

```python
titanic.groupby('age_category')['age_category'].count()
```

```python
# 10
age_count = titanic.groupby('age_category')['age_category'].count()
plt.pie(age_count, labels=age_count.index, autopct='%.1f%%', shadow=True)
plt.show()
```

```python
# 11
tot = titanic.groupby('age_category')['age_category'].count()
tot
surv = titanic[titanic['survived'] == 1].groupby('age_category')['age_category'].count()
surv
rate = surv / tot * 100
rate[rate == rate.max()]
```

```python
# 12
surv = pd.pivot_table(titanic, index=['sex', 'age_category', 'pclass'], aggfunc='sum')
surv
tot = pd.pivot_table(titanic, index=['sex', 'age_category', 'pclass'], aggfunc='count')
tot
result = surv
result['survival_rate'] = surv['survived'] / tot['survived'] * 100
result[['survival_rate']]
```

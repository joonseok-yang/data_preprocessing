```python
살인발생이 10건 이상이면 높음, 5건 이상이면 보통, 
나머지는 적음으로 표시하는 발생정도 컬럼을 추가하여 
(관서명, 살인발생,살인검거,발생정도) 를 출력하시요
```

```python
def fn( v ):
    if v>=10:
        return '높음'
    elif v>=5:
        return '보통'
    else:
        return '적음'
    
```

```python
df['살인발생정도']= df['살인 발생'].apply( fn )
df[['살인 발생','살인 검거','살인발생정도']]
```

```python
# 5. 강도발생이 높은 톱5를 구하시요 (관서명, 강도발생, 강도검거)
df.sort_values(by='강도 발생',
               ascending=False)[:5][['강도 발생','강도 검거']]
```

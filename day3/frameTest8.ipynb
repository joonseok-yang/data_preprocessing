```python
df.sort_values(by='남아', ascending=False)[:5]['남아']
# df['남아'].sort_values( ascending=False)[:5]
```

```python
# df['남아율'] = df['남아']/ (df['남아']+df['여아'])*100
df['남아율'] = df['남아']/ df.sum(axis=1)*100
```

```python
df.head()
```

```python
# 출생량 컬럼을 추가하고
# 남아수가 1800000 이상이면 '많음'아니면 '적음'
# 이라고 보여 주시요
b2000 = df.loc[2000:]
b2000['출생량'] = df.apply( lambda v: '많음' if v[0]>1800000 else '적음',
                        axis=1)
b2000
```

```python
# 9.1800년대,1900년대,2000년대 
# 각 남아,여야수 평균을 구하시요
b1800 = df.loc[1800:1899].mean()
b1900 = df.loc[1900:1999].mean()
b2000 = df.loc[2000:].mean()
# print( b1800)
# print( b1900)
# print( b2000)
pd.options.display.float_format = '{:.2f}'.format
my = pd.DataFrame( [b1800,b1900,b2000 ],
                  index=['1800년대','1900년대','2000년대'])
my
```

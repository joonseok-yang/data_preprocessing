```python
df = pd.read_csv(r'crime_in_Seoul.csv', engine='python', index_col='관서명', thousands=',')
df
```

```python
df[['절도 발생', '절도 검거']].plot(kind='bar', figsize=(20, 7))
plt.show()
```

```python
df['절도 검거율'] = df['절도 검거'] / (df['절도 발생']) * 100
df
```

```python
df['발생 정도'] = df.apply(lambda v: "높음" if v['살인 발생'] >= 10 else '보통' if v['살인 발생'] >= 5 else '낮음', axis=1)
df[['살인 발생', '살인 검거', '발생 정도']]
```

```python
def fn(v):
    if v>=10:
        return '높음'
    if v>=5:
        return '보통'
    else:
        return '낮음'
```

```python
df['발생 정도'] = df['살인 발생'].apply(fn)
df[['살인 발생', '살인 검거', '발생 정도']]
```
